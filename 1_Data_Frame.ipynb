# **Create Dataframe from List and Dictionary**

A DataFrame in Python is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It is a fundamental data structure in the Pandas library, widely used for data analysis and manipulation.

image.png

```python
#import library
import pandas as pd
#empty df
df=pd.DataFrame()
df
```

```python
#about df dataset
df.info()
```

**1D List to Dataframe**

```python
#create a list
mte=['CNC','Instrumentation','Mechanics','Circuit']
mte
```

```python
#check whether it is list or not
type(mte)
```

```python
#convert this list into Dataframe
df=pd.DataFrame(mte)
df
```

```python
#know the type of df dataset
type(df)
```

```python
#about row and column of df
row, col=df.shape
```

```python
row
```

```python
col
```

```python
#specify the column name of df
df=pd.DataFrame(mte, columns=['Mechatronics'])
df
```

```python
#set the row name
df=pd.DataFrame(mte, columns=['Mechatronics'], index=['c1','c2','c3','c4'])
df
```

```python
#Multiple lists?
rob=['DSP','DIP','AI','ML']
rob
```

```python
#combine multiple lists and convert into Dataframe
df2= pd.DataFrame(zip(mte,rob),columns=['Mechatronics', 'Robotics'],index=['c1','c2','c3','c4'])
df2
```

```python
#shape of df
df.shape
```

```python
#shape of df2
df2.shape
```

**2D list to Dataframe**

```python
#2D list
eee=[['DSP','DIP'],['SSP','PPE']]
eee
```

```python
#convert 2D list to Dataframe
df3=pd.DataFrame(eee)
df3
```

```python
df3=pd.DataFrame(eee,columns=['Communication','AP'])
df3
```

**Dictionary to Dataframe**

```python
#create a dictionary
dict1={
    'Mechatronics': mte,
    'Robotics': rob
}
dict1
```

```python
#convert the dictionary dict1 to Dataframe
df4=pd.DataFrame(dict1)
df4
```

```python
type(dict1)
```

```python
#multiple columns?
df['Date','Social Networking'] #returns errors, because it is list
```

```python
df[['Date','Social Networking']] #now it returns dataframe
```

```python
#sorting in ascending order
df['Social Networking'].sort_values()
```

```python
#sorting in descending order
df['Social Networking'].sort_values(ascending=False)
```

```python
#sorting the full dataframe
df.sort_values('Social Networking',ascending=False) #sort the full dataset with respect to Social Networking
```

**Statistical Analysis**

```python
#statistical description
df.describe()
```

```python
#correlation
df.corr(numeric_only=True)
```

```python
#correlation with Heatmap
import seaborn as sns
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
#subseting rows
df['Social Networking']>50
```

```python
#extract the values based on condition
df[df['Social Networking']>50]
```

```python
#subsetting rows by string values
df[df['Week Day']=='Wednesday']
```

```python
#Multiple conditions?
df[(df['Week Day']=='Wednesday')|(df['Week Day']=='Sunday')] #Logical OR condition
```

```python
#subsetting using isin()
df['Week Day'].isin(['Wednesday','Sunday'])
```

```python
df[df['Week Day'].isin(['Wednesday','Sunday'])]
```
