```python
#import libraries
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
df=pd.read_csv("credit.csv")
df.head()
```

```python
df.shape
```

**Profiling report generation**

Setting explorative=True enhances the profiling report by adding more detailed and advanced analyses, including:

*   Interactions between variables (pairwise scatter plots or heatmaps)
*   Correlations using multiple methods (Pearson, Spearman, Kendall, etc.)
*   Missing value visualizations
*   Sample data tables
*   Duplicate row detection
*   Time-series specific features (if datetime is detected)

When explorative=False (the default), the report is simpler and faster to generate.

```python
profile=ProfileReport(df,title='Pandas Profiling Report',explorative=True)
profile
```
