# **Merge two datasets to form large dataset**

The **concat()** function can be used to concatenate two Dataframes by adding the rows of one to the other. The **merge()** function is equivalent to the SQL JOIN clause. 'left', 'right' and 'inner' joins are all possible.

```python
import pandas as pd
#View dataset 1
df1=pd.read_csv('dataset1.csv')
df1
```

```python
#View dataset 2
df2=pd.read_csv('dataset2.csv')
df2
```

Method **pandas.merge()** and **DataFrame.merge()** are used to merge two or multiple DataFrames. Both these methods work exactly the same and they also take a similar number of params. Merging DataFrames is nothing but joining DataFrames similar to Database join.

```python
#Merge these two datasets into one
df3=pd.merge(df1,df2)
df3
```

```python
df3=pd.merge(df1,df2,on='SL') #same result will appear
df3
```

Similar to the Database join, merge() method also supports several join types like **left, right, inner, outer and cross.** Use how param to specify the join type.

By default merge() uses inner join on columns that are present on both DataFrames. You can also explicitly specify the columns you wanted to join on and join by row index.

Sometimes it is required to specify the column names you wanted to merge, to do so use on param. For example df3=pd.merge(df1,df2, on='Courses').

The “**outer**” merge combines all the rows for left and right dataframes with NaN when there are no matched values in the rows. Notice that the method indicator is set to True in order to indicate where each row originates from in the merge data — outer_merge.

```python
#union operation (intersection if inner)
df3=pd.merge(df1,df2,on='SL',how='outer') 
df3
```

To merge Pandas DataFrame, use the merge() function. In that, you can set the parameter indicator to True or False. If you want to check which dataframe has a specific record, then use − **indicator= True**. As shown above, using above parameter as True, adds a column to the output DataFrame called “_merge”.

```python
#check the avaiability of data in both datasets
df3=pd.merge(df1,df2,on='SL',how='outer',indicator= True) 
df3
```

```python
#check the avaiability of data in both datasets
df3=pd.merge(df1,df2,on='SL',how='right',indicator= True) 
df3
```
